# Distribution Stations

In nature, events follow different probability patterns. For example, what is the probability of getting heads when tossing a coin? What is the probability that an earthquake will occur within the next 10 years? If we vaccinate 200 people, what is the probability that at least 100 of them develop immunity? Since these events follow different patterns, statisticians have defined various *probability distributions* to model and predict occurrences. You may already be familiar with the *normal distribution*, but in statistics, there are several other important distributions.
